```python
# 1.0 Call libraries
import pandas as pd
import numpy as np
from  pathlib import Path

# 1.0.1
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# 2.1 Where is our data file:
pathToFolder = "/gdrive/MyDrive"
```

```python
# 2.2 Full file path:
path = Path(pathToFolder) / "mobile_price.csv"
```

```python
# 3.0 Read data
mobile_price = pd.read_csv(path)
mobile_price.head()
```

```python
# 3.1
mobile_price.shape
```

```python
mobile_price.dtypes
```

```python
# 3.2 Get target and predictors:
y = mobile_price.pop("mobile_wt")
X = mobile_price
```

```python
# 4.0 Build model:

model = tf.keras.Sequential()
```

```python
# 4.1
model.add(layers.Input( shape = (X.shape[1],) ) )
```

```python
# Adding Hidden Layers
model.add(layers.Dense(10,activation='relu'))
```

```python
model.add(layers.Dense(20,activation='relu'))
```

```python
model.add(layers.Dense(30,activation='relu'))
```

```python
model.add(layers.Dense(10,activation='relu'))
```

```python
model.compile( loss = "mae")
```

```python
model.fit(X,y, epochs = 500 )
```

```python
from tensorflow.keras.utils import plot_model
```

```python
plot_model(model, show_shapes=True)
```

```python
# 5.0 Predict
pred = model.predict(X)
```

```python
# 5.1 Round up decimal places
pred = np.round(pred)
```
